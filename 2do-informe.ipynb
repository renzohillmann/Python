```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy import optimize
import numpy as np
import math as math

plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"

#espectro del mercurio
with open("mercury1.txt") as sum:
    
    lines = sum.readlines()
    x = [float(line.split()[0]) for line in lines]
    y = [float(line.split()[1]) for line in lines]
    
x = np.array(x)
y = np.array(y)
# 
def gaussian(x, amplitude, mean, stddev):
    return amplitude * np.exp(-((x - mean) / 4 / stddev)**2)

def linear(x, m, b):
    return m*x+b

#popt, pcov = optimize.curve_fit(gaussian, x, y , [1, -22.5, 1] )

#print (popt)
fig = plt.figure(figsize=(15,9))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y, "darkviolet")

#axes.set_title("", fontsize=20)
axes.set_xlabel("$\lambda$ [nm]", fontsize=25)
axes.set_ylabel("Intensidad  [Unid.s arbitrarias]", fontsize=25)
axes.tick_params(axis="x", labelsize=20)
axes.tick_params(axis="y", labelsize=20)



plt.savefig("mercury1.pdf")
```
