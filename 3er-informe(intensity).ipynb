```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy import optimize
import numpy as np
import math as math

def linear (x, m, b):
    return m*x+b

def diode_function (x, A, B, C): #ecuaci√≥n del diodo de Shockley
    return A*(np.exp(B*(x - C)) - 1)

plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"

#espectro del mercurio
with open("I1.txt") as sum:
    
    lines = sum.readlines()
    x = [float(line.split()[0]) for line in lines]
    y = [float(line.split()[1]) for line in lines]
    
x = np.array(x)
y = np.array(y)
# 


#popt, pcov = optimize.curve_fit(gaussian, x, y , [1, -22.5, 1] )

#print (popt)
fig = plt.figure(figsize=(15,9))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y, "darkviolet", marker = "o")

#axes.set_title("", fontsize=20)
axes.set_xlabel("V [V]", fontsize=25)
axes.set_ylabel("I  [$\\rho$ A]", fontsize=25)
axes.tick_params(axis="x", labelsize=20)
axes.tick_params(axis="y", labelsize=20)

plt.axhline(y = -0.25)
plt.axhline(y = -0.7)
plt.axvline(x = -0.45)

plt.savefig("I1.pdf")
```
