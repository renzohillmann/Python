```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

import math

#parameters
w0 = 1
C = 2
Q1 = 1
Q2 = 1.5
Q3 = 2

# 100 linearly spaced numbers

w = w0*np.linspace(0.002,10,1000)

A1 = (C*w/w0)/np.sqrt((w0/w-w/w0)**2+1/(Q1**2))
A2 = (C*w/w0)/np.sqrt((w0/w-w/w0)**2+1/(Q2**2))
A3 = (C*w/w0)/np.sqrt((w0/w-w/w0)**2+1/(Q3**2))


fig = plt.figure(figsize=(15,12))
plt.title("Amplitudes para distintos valores de Q con C = 2.0", fontsize=20)
ax = fig.add_subplot(1,1,1)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_xlim([0,6])
ax.set_ylim([0,4.5])
plt.xlabel('$\omega$', fontsize=20)
plt.ylabel('Amplitud', fontsize = 20)
plt.plot(w, A1, color="blue")
plt.plot(w, A2, color="m")
plt.plot(w, A3, color="c")
plt.axhline(y=2, color='r', linestyle='--')
plt.savefig("AMPLITUD.pdf")
```
