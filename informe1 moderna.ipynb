```python
import matplotlib.pyplot as plt
from scipy import optimize
import numpy as np

with open("5kvd21.txt") as sum:
    
    lines = sum.readlines()
    x = [float(line.split()[0]) for line in lines]
    y = [float(line.split()[1]) for line in lines]
    
x = np.array(x)
y = np.array(y)

def gaussian(x, amplitude, mean, stddev):
    return amplitude * np.exp(-((x - mean) / 4 / stddev)**2)

popt, pcov = optimize.curve_fit(gaussian, x, y , [1, -22.5, 1] )

print (popt)
fig = plt.figure(figsize=(5,3))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    
axes.plot(x,y, "r")

axes.set_title("")
axes.set_xlabel("x")
axes.set_ylabel("Intensidad") #plt(x,gaussian(x, 1, 1, 1), "r")


fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    
axes.plot(x,gaussian(x, 14.92807002, -20.17531179, 0.5469663), "r")

axes.set_title("")
axes.set_xlabel("x")
axes.set_ylabel("Intensidad")

plt.savefig("5kVD21.pdf")


```
