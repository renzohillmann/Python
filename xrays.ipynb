```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy import optimize
import numpy as np
import math as math

def linear (x, m, b):
    return m*x+b

def child (x, k, a):
    return k*pow(x, a)

plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"

#NaCl crystal spectrum
with open("xray-data.txt") as sum:
    
    lines = sum.readlines()
    x1 = [float(line.split()[0]) for line in lines]
    y1 = [float(line.split()[1]) for line in lines]
    
x1 = np.array(x1)
y1 = np.array(y1)

#popt, pcov = optimize.curve_fit(gaussian, x, y , [1, -22.5, 1] )

#print (popt)
fig = plt.figure(figsize=(15,9))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
#axes.set_title("", fontsize=20)
axes.set_xlabel(r"$2\theta$ []", fontsize=25)
axes.set_ylabel(r"R [1/s]", fontsize=25)
axes.tick_params(axis="x", labelsize=20)
axes.tick_params(axis="y", labelsize=20)

axes.set_xlim([0, 25])


axes.plot(x1,y1, "blue", linewidth = 2) #, marker = "o", in case you need to add a marker

#labels
axes.text(7.5, 1500, r'1: $K_{\alpha}$', fontsize=20, color = "darkgreen")
axes.text(5, 820, r'1: $K_{\beta}$', fontsize=20, color = "darkgreen")

axes.text(14.8, 550, r'2: $K_{\alpha}$', fontsize=20, color = "darkgreen")
axes.text(11.5, 250, r'2: $K_{\beta}$', fontsize=20, color = "darkgreen")

axes.text(22.3, 150, r'3: $K_{\alpha}$', fontsize=20, color = "darkgreen")
axes.text(18, 80, r'3: $K_{\beta}$', fontsize=20, color = "darkgreen")

#1
plt.axvline(x = 6.45, linestyle = "dashed")
plt.axvline(x = 7.25, linestyle = "dashed")

#2
plt.axvline(x = 13, linestyle = "dashed")
plt.axvline(x = 14.6, linestyle = "dashed")

#3
plt.axvline(x = 19.65, linestyle = "dashed")
plt.axvline(x = 22.2, linestyle = "dashed")

plt.savefig("crystal-spectrum.pdf")
```
