```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import math

P = 1
E = 1
H = 2.0

#Original coordinates
x, y = np.meshgrid(np.linspace(-5,5, 20), np.linspace(-5,5, 20))


#Displacement field

def u(x, y):
    return (P/E)*x*y

def v(x, y):
    return (P/(2*E))*(x*x-y*y)

plt.quiver(x, y, u(x, y), v(x, y))
plt.show()
```

```python



#segment I
t1 =np.linspace(-H/2, H/2, 10)
x1 = t1
y1 = np.full((10), H/2)


#segment II
t2 =np.linspace(-H/2, H/2, 10)
x2 = np.full((10), H/2)
y2 = t2

#segment III
t3 =np.linspace(-H/2, H/2, 10)
x3 = t3
y3 = np.full((10), -H/2)

#segment IV
t4 =np.linspace(-H/2, H/2, 10)
x4 = np.full((10), -H/2)
y4 = t2



fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1,1,1)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
#ax.set_xlim([-3,3])
#ax.set_ylim([-3,3])
plt.plot(x1, y1, color="blue")
plt.plot(x2, y2, color="red")
plt.plot(x3, y3, color="orange")
plt.plot(x4, y4, color="magenta")


#displaced contour

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1,1,1)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_xlim([-3,3])
ax.set_ylim([-3,3])
plt.plot(x1+u(x1, y1), y1+v(x1, y1), color="blue")
plt.plot(x2+u(x2, y2), y2+v(x2, y2), color="red")
plt.plot(x3+u(x3, y3), y3+v(x3, y3), color="orange")
plt.plot(x4+u(x4, y4), y4+v(x4, y4), color="magenta")
plt.quiver(x1, y1, u(x1, y1), v(x1, y1))
plt.quiver(x2, y2, u(x2, y2), v(x2, y2))
plt.quiver(x3, y3, u(x3, y3), v(x3, y3))
plt.quiver(x4, y4, u(x4, y4), v(x4, y4))



```
