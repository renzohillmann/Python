```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_max = 50
res = 1000
x = np.linspace(-3*np.pi-1, +3*np.pi+1, res)

def b(n):
    return -2*np.cos(n*np.pi)/n


y = np.array([0]* res)

for n in range(1, N_max):
    y = y + b(n)*np.sin(n*x)

#plottinng the Fourier series:
plt.figure(figsize=(12, 6))

xticks_positions = [-3*np.pi, -2*np.pi, -1*np.pi, 0, 1*np.pi, 2*np.pi, 3*np.pi]
yticks_positions = [-np.pi, 0, np.pi]

plt.ylim(-np.pi -0.5, np.pi + 0.5)

xtick_labels = [r"$-3\pi$", r"$-2\pi$", r"$-1\pi$", r"$0$", r"$1\pi$", r"$2\pi$", r"$3\pi$"]
ytick_labels = [r"$-1\pi$", r"$0$", r"$1\pi$"]

plt.xticks(xticks_positions, xtick_labels, alpha=0.6)
plt.yticks(yticks_positions, ytick_labels, alpha=0.6)

plt.grid(True)

plt.plot(x,y)
plt.plot(x, x)
```
