```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

import math

#parameters
q = 0.1  #gamma/w0
F0 = 1
m = 0.1
k = 15.79
w0 = math.sqrt(k/m)
w02 = k/m


# 100 linearly spaced numbers

w = np.linspace(0,200,10000)

A1 = 10/np.sqrt((w02**2-w**2)**2+w02*w**2)
A2 = 10/np.sqrt((w02**2-w**2)**2+0.01*w02*w**2)
A3 = 10/np.sqrt((w02**2-w**2)**2+0.01**2*w02*w**2)

fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(1,1,1)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_xlim([110,200])
ax.set_ylim([0,0.007])
plt.xlabel('$\omega$', fontsize=20)
plt.ylabel('Amplitud', fontsize = 20)
plt.text(120, 0.006, r'$\gamma/\omega_0 = 1$', fontsize=20)
plt.plot(w,A1, color="blue")
plt.savefig("amplitud1.pdf")

fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(1,1,1)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_ylim([0,0.06])
ax.set_xlim([125,190])
plt.xlabel('$\omega$', fontsize=20)
plt.ylabel('Amplitud', fontsize = 20)
plt.text(130, 0.045, r'$\gamma/\omega_0 = 0.1$', fontsize=20)
plt.plot(w,A2, color="red")
plt.savefig("amplitud2.pdf")


fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(1,1,1)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_ylim([0,0.08])
ax.set_xlim([120,200])
plt.xlabel('$\omega$', fontsize=20)
plt.ylabel('Amplitud', fontsize = 20)
plt.text(130, 0.07, r'$\gamma/\omega_0 = 0.01$', fontsize=20)
plt.plot(w,A3, color="green")
plt.savefig("amplitud3.pdf")


fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(1,1,1)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_xlim([140,180])
ax.set_ylim([0,0.07])
plt.xlabel('$\omega$', fontsize=20)
plt.ylabel('Amplitud', fontsize = 20)
plt.plot(w,A1, color="blue")
plt.plot(w,A2, color="red")
plt.plot(w,A3, color="green")
plt.savefig("amplitud4.pdf")  

```

```python
E1 = 2.5*(w02+w**2)/((w02**2-w**2)**2+w02*w**2)
E2 = 2.5*(w02+w**2)/((w02**2-w**2)**2+0.01*w02*w**2)
E3 = 2.5*(w02+w**2)/((w02**2-w**2)**2+0.01**2*w02*w**2)

#Lorentzian functions
L1 = 1/((w0-w)**2+(w0/2)**2)
L2 = 1/((w0-w)**2+(0.1*w0/2)**2)
L3 = 1/((w0-w)**2+(0.01*w0/2)**2)



fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(1,1,1)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_xlim([110,200])
ax.set_ylim([0,0.04])
plt.xlabel('$\omega$', fontsize=20)
plt.ylabel('<E>', fontsize = 20)
plt.text(120, 0.035, r'$\dfrac{\gamma}{\omega_0} = 1$', fontsize=20)
plt.plot(w,E1, color="blue")
plt.savefig("energia1.pdf")  


fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(1,1,1)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_ylim([0,1.7])
ax.set_xlim([125,190])
plt.xlabel('$\omega$', fontsize=20)
plt.ylabel('<E>', fontsize = 20)
plt.text(130, 1.4, r'$\dfrac{\gamma}{\omega_0} = 0.1$', fontsize=20)
plt.plot(w,E2, color="red")
plt.savefig("energia2.pdf")  


fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(1,1,1)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_ylim([0,3])
ax.set_xlim([120,200])
plt.xlabel('$\omega$', fontsize=20)
plt.ylabel('<E>', fontsize = 20)
plt.text(130, 2.5, r'$\dfrac{\gamma}{\omega_0} = 0.01$', fontsize=20)
plt.plot(w,E3, color="green")
plt.savefig("energia3.pdf")  

fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(1,1,1)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_xlim([145,170])
ax.set_ylim([0,2])
plt.xlabel('$\omega$', fontsize=20)
plt.ylabel('<E>', fontsize = 20)
plt.plot(w,E1, color="blue")
plt.plot(w,E2, color="red")
plt.plot(w,E3, color="green")
plt.savefig("energia4.pdf")  






```
