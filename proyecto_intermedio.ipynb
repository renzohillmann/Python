```python
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
import matplotlib.cm as cm
import matplotlib.patches as mpatches



data = np.random.rand(10, 10)*20
#data = 
# create discrete colormap
cmap = colors.ListedColormap(['white', 'black'])
bounds = [0,10,20]
norm = colors.BoundaryNorm(bounds, cmap.N)

fig, ax = plt.subplots()
ax.imshow(data, cmap=cmap, norm=norm)

# draw gridlines
#ax.grid(which='major', axis='both', linestyle='-', color='k', linewidth=2)
ax.set_xticks(np.arange(-.5, 10, 1));
ax.set_yticks(np.arange(-.5, 10, 1));

plt.show()
```

```python
N = 15
# make an empty data set
data = np.ones((N, N)) * np.nan
# fill in some fake data
for j in range(3)[::-1]:
    data[N//2 - j : N//2 + j +1, N//2 - j : N//2 + j +1] = j
# make a figure + axes
fig, ax = plt.subplots(1, 1, tight_layout=True)
# make color map
my_cmap = colors.ListedColormap(['r', 'g', 'b'])
# set the 'bad' values (nan) to be white and transparent
my_cmap.set_bad(color='w', alpha=0)
# draw the grid
for x in range(N + 1):
    ax.axhline(x, lw=2, color='k', zorder=5)
    ax.axvline(x, lw=2, color='k', zorder=5)
# draw the boxes
ax.imshow(data, interpolation='none', cmap=my_cmap, extent=[0, N, 0, N], zorder=0)
# turn off the axis labels
ax.axis('off')
```

```python
x = [32, 64, 128, 256, 512]
y = [0.175, 0.649, 2.553, 10.320, 44.996] #O2
z = [0.164, 0.622, 2.480, 9.861, 43.986] #O3

plt.plot(x,y, "m-.", x,z, "c:") #markers!
plt.xlabel("Número de columnas de la matriz (L)")
plt.ylabel("Tiempo de ejecución (segs)")
plt.title("O2 vs O3", color="#34495e")
high_legend = mpatches.Patch(color="magenta", label="O2")
low_legend = mpatches.Patch(color="cyan", label="O3")
plt.legend(handles=[high_legend,low_legend])
plt.savefig("tiempos.pdf", dpi = 200)
```
