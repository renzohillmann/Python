```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as opt
import numpy as np
```

```python
def parabola(x, a, b, c):
    return a * x**2 + b * x + c

def linear(x, m, b):
    return m * x + b
```

```python
# Load data from the Excel file

try:
  df_balon1 = pd.read_excel("datos tracker.xlsx", skiprows=1)
except FileNotFoundError:
  print("Error: datos tracker.xlsx' not found.")
  exit()
  
try:
  df_balon = pd.read_excel("balon tracker2.xlsx", skiprows=1)
except FileNotFoundError:
  print("Error: 'balon tracker2.xlsx' not found.")
  exit()

try: 
  df_icopor = pd.read_excel("icopor tracker2.xlsx", skiprows=1)
except FileNotFoundError:
  print("Error: 'icopor tracker2.xlsx' not found.")
  exit()
```

# 1) $y$ vs $x$

```python

# Adapt column names if different
x_balon = df_balon['x']
y_balon = df_balon['y']

x_icopor = df_icopor['x']
y_icopor = df_icopor['y']

#define the function for the regression curve

popt, pcov = opt.curve_fit(parabola, x_balon, y_balon)
print("Parabola coefficients for balon:", popt)

plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"

# Add the new set of points to the existing plot
plt.plot(x_balon, y_balon, marker='o', linestyle='None', color='r', label='Balón')
plt.plot(x_icopor, y_icopor, marker = 'o', linestyle='None', color='b', label='Icopor')
# Plot the fitted curve
x_fit = np.linspace(min(x_balon), max(x_balon), 100)
y_fit = parabola(x_fit, *popt)
plt.plot(x_fit, y_fit, color='r', label='Ajuste parabólico')

# Add labels, title, legend (if needed)
plt.xlabel("Posición en x (m)")
plt.ylabel("Posición en y (m)")
#plt.title("Plot with Data from Two Sources")
plt.legend()  # Show the legend

# image size
plt.gcf().set_size_inches(6, 4)
# Display the plot
plt.show()
```

```python
# Adapt column names if different
x_balon = df_balon['x']
y_balon = df_balon['y']

x_icopor = df_icopor['x']
y_icopor = df_icopor['y']

#define the function for the regression curve

def parabola(x, a, b, c):
    return a * x**2 + b * x + c

popt, pcov = opt.curve_fit(parabola, x_balon, y_balon)
print("Parabola coefficients for balon:", popt)

plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"

# Add the new set of points to the existing plot
plt.plot(x_balon, y_balon, marker='o', linestyle='None', color='royalblue', label='Balón')
plt.plot(x_icopor, y_icopor, marker = 'o', linestyle='None', color='orange', label='Icopor')
# Plot the fitted curve
x_fit = np.linspace(min(x_balon), max(x_balon), 100)
y_fit = parabola(x_fit, *popt)
plt.plot(x_fit, y_fit, color='r', label='Ajuste parabólico', linewidth='2.5')

# Add labels, title, legend (if needed)
plt.xlabel("Posición en $x$ (m)")
plt.ylabel("Posición en $y$ (m)")
#plt.title("Plot with Data from Two Sources")
plt.legend()  # Show the legend

# image size
plt.gcf().set_size_inches(6, 4)
plt.savefig("y-vs-x.png", dpi=300, bbox_inches="tight")
# Display the plot
plt.show()
```

# 2) $y$ y $x$ vs $t$ 

```python

# Adapt column names if different
t_balon = df_balon['t']
x_balon = df_balon['x']

t_icopor = df_icopor['t']
x_icopor = df_icopor['x']

#define the function for the regression curve



popt, pcov = opt.curve_fit(linear, t_balon, x_balon, [1,1])
print("Parabola coefficients for balon:", popt)

plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"

# Add the new set of points to the existing plot
plt.plot(t_balon, x_balon, marker='o', linestyle='None', color='royalblue', label='Balón')
plt.plot(t_icopor, x_icopor, marker = 'o', linestyle='None', color='orange', label='Icopor')
# Plot the fitted curve
t_fit = np.linspace(min(t_balon), max(t_balon), 100)
x_fit = linear(t_fit, *popt)
plt.plot(t_fit, x_fit, color='r', label='Ajuste lineal', linewidth='2.5')

# Add labels, title, legend (if needed)
plt.xlabel("Tiempo (segundos)")
plt.ylabel("Posición en $x$ (metros)")
plt.legend()  # Show the legend

# image size
plt.gcf().set_size_inches(6, 4)
plt.savefig("x-vs-t.png", dpi=300, bbox_inches="tight")
# Display the plot
plt.show()
```
