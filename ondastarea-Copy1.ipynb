```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

import math

#parameters
q = 0.1  #gamma/w0
F0 = 1
m = 0.1
k = 15.79
w0 = math.sqrt(k/m)
w02 = k/m


# 100 linearly spaced numbers

w = np.linspace(0,200,10000)

A1 = 10/np.sqrt((w02**2-w**2)**2+w02*w**2)
A2 = 10/np.sqrt((w02**2-w**2)**2+0.01*w02*w**2)
A3 = 10/np.sqrt((w02**2-w**2)**2+0.01**2*w02*w**2)

fig = plt.figure(figsize=(20,15))
#plt.title("Amplitudes para distintos valores de $\gamma / \omega_0$", fontsize=20)

ax = fig.add_subplot(2,2,1)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_xlim([110,200])
ax.set_ylim([0,0.01])
plt.xlabel('frecuencia angular $\omega$', fontsize=16)
plt.ylabel('Amplitud (m)', fontsize = 16)
plt.text(175, 0.008, r'$\dfrac{\gamma}{\omega_0} = 1$', fontsize=20)
plt.plot(w,A1, color="#380282", linewidth=2)
#ax.set_title("Title for first plot",fontsize = 16)


ax = fig.add_subplot(2,2,2)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_ylim([0,0.06])
ax.set_xlim([125,190])
plt.xlabel('frecuencia angular $\omega$', fontsize=16)
plt.ylabel('Amplitud (m)', fontsize = 16)
plt.text(170, 0.045, r'$\dfrac{\gamma}{\omega_0} = 0.1$', fontsize=20)
plt.plot(w,A2, color="#ff796c", linewidth=2)

ax = fig.add_subplot(2,2,3)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_ylim([0,0.1])
ax.set_xlim([120,200])
plt.xlabel('frecuencia angular $\omega$', fontsize=16)
plt.ylabel('Amplitud (m)', fontsize = 16)
plt.text(175, 0.07, r'$\dfrac{\gamma}{\omega_0} = 0.01$', fontsize=20)
plt.plot(w,A3, color="#06c2ac", linewidth=2)

ax = fig.add_subplot(2,2,4)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_xlim([140,180])
ax.set_ylim([0,0.07])
plt.xlabel('frecuencia angular $\omega$', fontsize=16)
plt.ylabel('Amplitud (m)', fontsize = 16)
plt.plot(w,A1, color="#380282", linewidth=2)
plt.plot(w,A2, color="#ff796c", linewidth=2)
plt.plot(w,A3, color="#06c2ac", linewidth=2)


plt.savefig("Amplitudes.pdf")  

```

```python
E1 = 2.5*(w02+w**2)/((w02**2-w**2)**2+w02*w**2)
E2 = 2.5*(w02+w**2)/((w02**2-w**2)**2+0.01*w02*w**2)
E3 = 2.5*(w02+w**2)/((w02**2-w**2)**2+0.01**2*w02*w**2)

#Lorentzian functions
L1 = 1/((w0-w)**2+(w0/2)**2)
L2 = 1/((w0-w)**2+(0.1*w0/2)**2)
L3 = 1/((w0-w)**2+(0.01*w0/2)**2)


fig = plt.figure(figsize=(20,15))
#plt.title("Amplitudes para distintos valores de $\gamma / \omega_0$", fontsize=20)


ax = fig.add_subplot(2,2,1)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_xlim([110,200])
ax.set_ylim([0,0.05])
plt.xlabel('frecuencia angular $\omega$', fontsize=16)
plt.ylabel('Energia promedio <E>', fontsize = 16)
plt.text(175, 0.04, r'$\dfrac{\gamma}{\omega_0} = 1$', fontsize=20)
plt.plot(w,E1, color="#380282", linewidth=2)

#ax.set_title("Title for first plot",fontsize = 16)


ax = fig.add_subplot(2,2,2)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_ylim([0,1.8])
ax.set_xlim([125,190])
plt.xlabel('frecuencia angular $\omega$', fontsize=16)
plt.ylabel('Energia promedio <E>', fontsize = 16)
plt.text(170, 1.4, r'$\dfrac{\gamma}{\omega_0} = 0.1$', fontsize=20)
plt.plot(w,E2, color="#ff796c", linewidth=2)

ax = fig.add_subplot(2,2,3)
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_ylim([0,3])
ax.set_xlim([120,200])
plt.xlabel('frecuencia angular $\omega$', fontsize=16)
plt.ylabel('Energia promedio <E>', fontsize = 16)
plt.text(175, 2.5, r'$\dfrac{\gamma}{\omega_0} = 0.01$', fontsize=20)
plt.plot(w,E3, color="#06c2ac", linewidth=2)

ax = fig.add_subplot(2,2,4)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.set_xlim([140,180])
ax.set_ylim([0,2])
plt.xlabel('frecuencia angular $\omega$', fontsize=16)
plt.ylabel('Energia promedio <E>', fontsize = 16)
plt.plot(w,E1, color="#380282", linewidth=2)
plt.plot(w,E2, color="#ff796c", linewidth=2)
plt.plot(w,E3, color="#06c2ac", linewidth=2)






plt.savefig("eyl.pdf")  
```
