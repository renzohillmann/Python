```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

import math

#parameters
Q = 2  # Q= ω0/γ
F0 = 1
γ = 1
ω0 = γ*Q


# 100 linearly spaced numbers

ω = np.linspace(0,5,1000)

A1 = 10/np.sqrt((ω0**2-ω**2)**2+ (γ*ω)**2)
#A2 = 10/np.sqrt((ω2**2-ω**2)**2+0.01*ω2*w**2)
#A3 = 10/np.sqrt((ω2**2-ω**2)**2+0.01**2*ω2*w**2)

fig = plt.figure(figsize=(10,7))
#plt.title("Amplitudes para distintos valores de $\gamma / \omega_0$", fontsize=20)

ax = fig.add_subplot(2,2,1)
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
#ax.set_xlim([0,5])
ax.set_ylim([0,6])
plt.xlabel('angular frequency $\omega$', fontsize=10)
plt.ylabel('Amplitude A(ω)', fontsize = 10)
#plt.text(175, 0.008, r'$\dfrac{\gamma}{\omega_0} = 1$', fontsize=20)
plt.plot(ω,A1, color="orange", linewidth=3)
plt.vlines(ω0, -1, 100, color = 'blue', lw = 1.5, alpha = 1, ls = 'dotted')


#adding labels
plt.text(ω0, 0, r'$ω_0$', color = 'b', ha='left', va='bottom')
#ax.set_title("Title for first plot",fontsize = 16)

plt.savefig("amplitude.pdf", bbox_inches='tight')
```
